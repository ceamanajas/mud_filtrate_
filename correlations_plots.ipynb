```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
corr1 = np.loadtxt('/content/drive/MyDrive/doutorado/mud_filtrate/dados/corr_70_10_20.txt')
corr2 = np.loadtxt('/content/drive/MyDrive/doutorado/mud_filtrate/dados/corr_70_20_20.txt')
corr3 = np.loadtxt('/content/drive/MyDrive/doutorado/mud_filtrate/dados/corr_70_30_20.txt')
corr4 = np.loadtxt('/content/drive/MyDrive/doutorado/mud_filtrate/dados/corr_70_35_20.txt')
```

```python
# saturations values
sat = [0, 5, 20, 40, 60, 100]

fig_corr = go.Figure()

fig_corr.add_trace(go.Scatter(x=sat, y=corr1, mode='lines+markers', name='10% of porosity',
                              line=dict(color='blue'), marker=dict(size=10)))
fig_corr.add_trace(go.Scatter(x=sat, y=corr2, mode='lines+markers', name='20% of porosity',
                              line=dict(color='green'), marker=dict(size=10)))
fig_corr.add_trace(go.Scatter(x=sat, y=corr3, mode='lines+markers', name='30% of porosity',
                              line=dict(color='violet'), marker=dict(size=10)))
fig_corr.add_trace(go.Scatter(x=sat, y=corr4, mode='lines+markers', name='35% of porosity',
                              line=dict(color='red'), marker=dict(size=10)))

# axis properties
fig_corr.update_xaxes(title_text='Mud filtrate saturation (%)',
                 title_font=dict(size=22), tickfont=dict(size=20))
fig_corr.update_yaxes(title_text='Pearson Correlation Coefficient',
                 title_font=dict(size=22), tickfont=dict(size=20))

# figure layout
fig_corr.update_layout(title='Well tie correlations - reservoir 70m thick',
                       title_font=dict(size=24),
                       xaxis=dict(tickfont=dict(size=20)),
                       yaxis=dict(tickfont=dict(size=20)),
                       legend=dict(font=dict(size=20)),
                       height=500,
                       width=800,
                       title_x=0.5)

fig_corr.show()
```
