```python
# saturations values
sat = [0, 5, 20, 40, 60, 100]

fig = make_subplots(rows=1, cols=2, column_widths=[0.2, 0.001], subplot_titles=(''),
                    horizontal_spacing=0.12, vertical_spacing=0.2)

fig.add_trace(go.Scatter(x=sat, y=corr1, mode='lines+markers', name='10%',
                              line=dict(color='#3366CC'), marker=dict(size=10)), row=1, col=1)
fig.add_trace(go.Scatter(x=sat, y=corr2, mode='lines+markers', name='15%',
                              line=dict(color='#DC3912'), marker=dict(size=10)), row=1, col=1)
fig.add_trace(go.Scatter(x=sat, y=corr3, mode='lines+markers', name='20%',
                              line=dict(color='#109618'), marker=dict(size=10)), row=1, col=1)
fig.add_trace(go.Scatter(x=sat, y=corr4, mode='lines+markers', name='30%',
                              line=dict(color='#FF9900'), marker=dict(size=10)), row=1, col=1)
fig.add_trace(go.Scatter(x=sat, y=corr5, mode='lines+markers', name='35%',
                              line=dict(color='#990999'), marker=dict(size=10)), row=1, col=1)

# Propriedades dos eixos
fig.update_xaxes(title_text='Mud filtrate saturation (%)',
                      title_font=dict(size=22), tickfont=dict(size=20), row=1, col=1)
fig.update_yaxes(title_text='Pearson Correlation Coefficient',
                      title_font=dict(size=22), tickfont=dict(size=20), row=1, col=1)

# Layout da figura
fig.update_layout(height=400, width=600, showlegend=True, title_text='<b>Well-to-seismic tie (70m thick)<b>', title_font=dict(size=24), title_x=0.5)

fig.update_annotations(font=dict(size=24))

# Update layout for legend
fig.update_layout(
    legend=dict(
        x=0.95,
        y=0.5,
        traceorder='normal',
        font=dict(size=20),
        bordercolor="Black",
        title=dict(text="Porosity", font=dict(size=24))  # Add legend title
    )
)

fig.show()
```

```python
filtrate_sat = np.array([0, 5, 20, 40, 60, 100])

# Dados a serem plotados
kdry_data = [kdry10_, kdry15_, kdry20_, kdry30_, kdry35_]
ksat_data = [ksat10_, ksat15_, ksat20_, ksat30_, ksat35_]
rhofl_data = [rhofl10_, rhofl15_, rhofl20_, rhofl30_, rhofl35_]
rho_data = [rho10_, rho15_, rho20_, rho30_, rho35_]
vp_data = [vp10_, vp15_, vp20_, vp30_, vp35_]
vs_data = [vs10_, vs15_, vs20_, vs30_, vs35_]

# Labels para as legendas
labels = ['10', '15', '20', '30', '35']
colors = ['#3366CC', '#DC3912', '#109618', '#FF9900', '#990999']
legends = ['10%', '15%', '20%', '30%', '35%']

# Criando subplots
fig = make_subplots(rows=2, cols=4, column_widths=[0.2, 0.2, 0.2, 0.001], subplot_titles=('Dry rock bulk modulus', 'Fluid density', 'P-wave velocity', '', 'Rock bulk modulus', 'Density', 'S-wave velocity'),
                    horizontal_spacing=0.12, vertical_spacing=0.2)

# Plotando os dados em subplots
for y_data, label, color, legend in zip(kdry_data, labels, colors, legends):
    fig.add_trace(go.Scatter(x=filtrate_sat, y=y_data, mode='lines+markers', name=legend, marker=dict(color=color)), row=1, col=1)

for y_data, label, color in zip(ksat_data, labels, colors):
    fig.add_trace(go.Scatter(x=filtrate_sat, y=y_data, mode='lines+markers', name=legend, marker=dict(color=color), showlegend=False), row=2, col=1)

for y_data, label, color in zip(rhofl_data, labels, colors):
    fig.add_trace(go.Scatter(x=filtrate_sat, y=y_data, mode='lines+markers', name=legend, marker=dict(color=color), showlegend=False), row=1, col=2)

for y_data, label, color in zip(rho_data, labels, colors):
    fig.add_trace(go.Scatter(x=filtrate_sat, y=y_data, mode='lines+markers', name=legend, marker=dict(color=color), showlegend=False), row=2, col=2)

for y_data, label, color in zip(vp_data, labels, colors):
    fig.add_trace(go.Scatter(x=filtrate_sat, y=y_data, mode='lines+markers', name=legend, marker=dict(color=color), showlegend=False), row=1, col=3)

for y_data, label, color in zip(vs_data, labels, colors):
    fig.add_trace(go.Scatter(x=filtrate_sat, y=y_data, mode='lines+markers', name=legend, marker=dict(color=color), showlegend=False), row=2, col=3)

# Propriedades do eixo x
fig.update_xaxes(title_text='', row=1, col=1, title_font=dict(size=22), tickfont=dict(size=20), showticklabels=False)
fig.update_xaxes(title_text='Filtrate saturation (%)', row=2, col=1, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_xaxes(title_text='', row=1, col=2, title_font=dict(size=22), tickfont=dict(size=20), showticklabels=False)
fig.update_xaxes(title_text='Filtrate saturation (%)', row=2, col=2, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_xaxes(title_text='', row=1, col=3, title_font=dict(size=22), tickfont=dict(size=20), showticklabels=False)
fig.update_xaxes(title_text='Filtrate saturation (%)', row=2, col=3, title_font=dict(size=22), tickfont=dict(size=20))

# Propriedades do eixo y
fig.update_yaxes(title_text='(GPa)', row=1, col=1, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_yaxes(title_text='(GPa)', row=2, col=1, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_yaxes(title_text='(g/cm³)', row=1, col=2, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_yaxes(title_text='(g/cm³)', row=2, col=2, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_yaxes(title_text='(km/s)', row=1, col=3, title_font=dict(size=22), tickfont=dict(size=20))
fig.update_yaxes(title_text='(km/s)', row=2, col=3, title_font=dict(size=22), tickfont=dict(size=20))

# Layout da figura
fig.update_layout(height=500, width=1200, showlegend=True, title_text='<b>Fluid substitution parameters (70m thick)<b>', title_font=dict(size=24), title_x=0.5)

fig.update_annotations(font=dict(size=24))

# Update layout for legend
fig.update_layout(
    legend=dict(
        x=0.95,
        y=0.5,
        traceorder='normal',
        font=dict(size=20),
        bordercolor="Black",
        title=dict(text="Porosity", font=dict(size=24))  # Add legend title
    )
)

fig.show()
```
